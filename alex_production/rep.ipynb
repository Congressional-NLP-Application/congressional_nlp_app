# Senate Data Pull: Twitter Handles

Using data provided through GitHub, I found a csv with all current senator information on it. Pulled and grabbed twitter handles. For future application use, we will want to either find a way to automate updates on the list or manually adjust after elections are held.

```python
import pandas as pd
import tweepy
```

```python
# Created df with data, then filtered to another df the data I wanted.
df = pd.read_csv('https://raw.githubusercontent.com/CivilServiceUSA/us-senate/master/us-senate/data/us-senate.csv')
df2 = df[['name','state_code', 'twitter_handle']]
df2
```

```python
# Created a handles only list
handles = df2.twitter_handle
handles.head()
```

```python
# Verify list is iterable.
for handle in handles:
    print(handle)
```

```python
# Create csv to read into .py file
handles.to_csv('twitter_handles.csv')
```
